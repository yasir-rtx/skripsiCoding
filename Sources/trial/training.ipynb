```python
cap = cv2.VideoCapture(1)
sum = 0

# messagebox.showinfo("Detail Pose Pertama", "Deskripsi Pose\nTekan enter untuk memulai")
print("Deskripsi Pose\nTekan enter untuk memulai")
while 1:
    # time.sleep(1)
    _, liveView = cap.read()
    face = HaarCascade.detectMultiScale(liveView, 1.3, 5)
    if len(face) > 0:
        x1, y1, w, h = face[0]
    else :
        x1, y1, w, h = 1,1,10,10
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1+w, y1+h
    cv2.rectangle(liveView, (x1,y1), (x2,y2), (0,255,0), 2)
    
    key = cv2.waitKey(5) & 0xFF
    if (sum == 0):
        desc = "front"
    elif (sum ==1):
        desc = "side"
    else:
        desc = "glass"
    if key == 13:
        for i in range(10):
            time.sleep(0.2)
            _, imgVideo = cap.read()
            
            FaceDetect = HaarCascade.detectMultiScale(imgVideo, 1.3, 5)
            if len(FaceDetect) > 0:
                x1, y1, w, h = FaceDetect[0]
            else :
                x1, y1, w, h = 1,1,10,10
                
            x1, y1 = abs(x1), abs(y1)
            x2, y2 = x1+w, y1+h
            
            img = cv2.cvtColor(imgVideo, cv2.COLOR_RGB2BGR)
            img = Image.fromarray(img)
            img_array = asarray(img)
            
            face = img_array[y1:y2, x1:x2]
            face = Image.fromarray(face)
            face = face.convert('RGB')
            face = face.resize((160,160))
            # face = asarray(face)
            
            
            fileName = labelPath + name + "_" + desc + "_" + str(i) + ".jpg"
            # cv2.imwrite(fileName, img_array)
            face.save(fileName)
            print(f"Menyimpan File : \"{fileName}\"")
        sum+=1
        
    cv2.imshow('res', liveView)
    if sum == 3:
        break
    if key == 27:
        break

    
cv2.destroyAllWindows()
cap.release()
```
