```python
from keras.models import load_model
from PIL import Image
from numpy import savez_compressed
from numpy import expand_dims
from numpy import asarray
from numpy import load
from os import listdir, mkdir
from os.path import isdir, exists
import cv2
import time
import pickle
```

```python
MyFaceNet = load_model("../../Models/FaceNet/facenet_keras.h5")
HaarCascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))
savePath = "..\\..\\..\\Dataset\\Dataset\\"
if not exists(savePath):
    mkdir(savePath)

label = "18101152630034"
label = label + "\\"
name = "gifra"
    
labelPath = savePath + label
print(labelPath)
if not exists(labelPath):
    mkdir(labelPath)
else:
    print("label sudah ada")


signature_path = "../../Data/"

if len(listdir(signature_path)) > 0:
    mySignature = open("../../Data/signature.pkl", "rb")
    faceDatabase = pickle.load(mySignature)
    mySignature.close()
else:
    faceDatabase = {}
```

```python
cap = cv2.VideoCapture(1)
sum = 0

# messagebox.showinfo("Detail Pose Pertama", "Deskripsi Pose\nTekan enter untuk memulai")
print("Deskripsi Pose\nTekan enter untuk memulai")
while 1:
    key = cv2.waitKey(5) & 0xFF
    _, imgVideo = cap.read()
    
    desc = "front" if sum<10 else "side" if sum<20 else "glass"
    # if (sum == 0):
    #     desc = "front"
    # elif (sum ==1):
    #     desc = "side"
    # else:
    #     desc = "glass"
    
    FaceDetect = HaarCascade.detectMultiScale(imgVideo, 1.3, 5)
    if len(FaceDetect) > 0:
        x1, y1, w, h = FaceDetect[0]
    else :
        x1, y1, w, h = 1,1,10,10
        
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1+w, y1+h
    
    img = cv2.cvtColor(imgVideo, cv2.COLOR_BGR2RGB)
    img = Image.fromarray(img)
    img_array = asarray(img)
    
    face = img_array[y1:y2, x1:x2]
    
    face = Image.fromarray(face)
    face_save = face.resize((160,160))
    face = asarray(face_save)
    
    if key == 13:
        fileName = labelPath + name + "_" + str(sum) + "_" + desc + ".jpg"
        print(f"saving : {fileName}")
        face_save.save(fileName)
        
        face = face.astype('float32')
        mean, std = face.mean(), face.std()
        face = (face - mean) / std
        
        face = expand_dims(face, axis=0)
        signature = MyFaceNet.predict(face)
        faceDatabase[name] = signature
        
        sum+=1
    
    cv2.rectangle(imgVideo, (x1,y1), (x2,y2), (0,255,0), 2)
    cv2.imshow('Rekam Wajah', imgVideo)
    if sum == 30:
        break
    if key == 27:
        break

cv2.destroyAllWindows()
cap.release()

mySignature = open("../../Data/signature.pkl", "wb")
pickle.dump(faceDatabase, mySignature)
mySignature.close()
```
